```python
def correlation_ratio(categories, measurements):
    fcat, _ = pd.factorize(categories)
    cat_num = np.max(fcat)+1
    y_avg_array = np.zeros(cat_num)
    n_array = np.zeros(cat_num)
    for i in range(0,cat_num):
        cat_measures = measurements[np.argwhere(fcat == i).flatten()]
        n_array[i] = len(cat_measures)
        y_avg_array[i] = np.average(cat_measures)
    y_total_avg = np.sum(np.multiply(y_avg_array,n_array))/np.sum(n_array)
    numerator = np.sum(np.multiply(n_array,np.power(np.subtract(y_avg_array,y_total_avg),2)))
    denominator = np.sum(np.power(np.subtract(measurements,y_total_avg),2))
    if numerator == 0:
        eta = 0.0
    else:
        eta = np.sqrt(numerator/denominator)
    return eta
```

### Inspecting if our residual following the normal distribution or not using Shapiro test

```python
def try_normal(data, residual):
    result = {'Shapiro' : shapiro(data)[1],
              'Skewness': skew(residual),
              'Kurtosis': kurtosis(residual)
    }
    return result
```

```python
try_normal(data, result.resid)
```

Because the shapiro result test from result.resid is 0 and it is below than alpha = 0.05, so we can conclude that the residual from result1 is not Normal Distribution

### Checking for 'Autocorrelation' using Ljungbox test

- H0: Autocorrelation is absent
- H1: Autocorrelation is present

HO is accepted if alpha < p-value

```python
from statsmodels.stats import diagnostic
```

```python
# Autocorrelation checking for result.resid

diagnostic.acorr_ljungbox(result.resid, lags = 1)
#output (stat, p-value)
```

Because p-value for result.resid is greater than alpha = 0.05, so for those target we can conclude that each residual is autocorrelation.

### Checking for Heteroscedascity using Goldfeld Quandt test
- H0: Error terms are homoscedastic
- H1: Erorr terms are Heteroscedascity

HO is accepted if alpha < p-value

```python
import statsmodels.stats.api as sms
from statsmodels.compat import lzip
```

```python
# Heteroscedascity checking for result.resid

name = ['F statistic', 'p-value']
test = sms.het_goldfeldquandt(result.resid, result.model.exog)
lzip(name, test)
```

Because p-value for result.resid is below than alpha = 0.05, so for those target we can conclude that each residual is Homoscedastic.

```python
def calculate_model(publisher):
    coef_var = {}
    
    data_new2 = new_steam[new_steam['publisher'] == publisher]
    target3 = data_new2['max_owners']
    
    x_train_new2, x_test_new2, y_train_new2, y_test_new2 = train_test_split(data_new2.drop(columns=['publisher', 'max_owners']), target3, test_size = 0.3, random_state = 101)
    
    lm_steam_new2 = LinearRegression()
    lm_steam_new2.fit(x_train_new2, y_train_new2)
    
    coef_var['intercept'] = lm_steam_new2.intercept_
    
    for i in range(len(x_train_new2.columns)):
        coef_var[x_train_new2.columns[i]] = lm_steam_new2.coef_[i]
            
    return coef_var
```

```python
def get_model(publisher):
    coef_var = calculate_model(publisher)
    y_formula = ''
    
    for i,j in zip(coef_var.keys(), coef_var.values()):
        if (i == 'intercept'):
            y_formula = 'Y = {:.2f}'.format(j)
        else:
            if str(j)[0] == '-':
                y_formula += ' - {:.2f}*X_{}'.format(float(str(j)[1:]), i)
            else:
                y_formula += ' + {:.2f}*X_{}'.format(j,i)
    
    print(y_formula)
```

```python
from scipy import stats
import math
```

```python
def est_param(mean_x, std_x, len_x, ci):
    if len_x < 30:
        critical_value = stats.t.ppf(ci, len_x-1)
    else:
        critical_value = stats.norm.ppf(ci)

    margin_of_error = critical_value*(std_x/math.sqrt(len_x))
    lower = mean_x-margin_of_error
    upper = mean_x+margin_of_error
    
    return [int(round(lower)), int(round(upper))]
```

```python
def bootstrap(n_size, n_boot, ci, data):
#     n_size = len(data)
    mean_list = list()
    
    for i in range(n_boot):
        resamp_data = data.sample(n = n_size, replace=True)
        mean_resamp = resamp_data.mean()
        mean_list.append(mean_resamp)
    
    a = (1.0-ci)/2.0
    
    #confidence interval
    p1 = (a)*100
    lower = np.percentile(mean_list,p1)
    p2 = (ci+(a))*100
    upper = np.percentile(mean_list,p2)
    
    return [int(round(lower)), int(round(upper))]
```

```python
def recommended_val(publisher, ci):
    data_x3 = new_steam[new_steam['publisher'] == publisher]
    data_x3 = data_x3.drop(columns=['publisher', 'max_owners'])
    recommended_low = []
    recommended_up = []
    recommended_dict = {}
    
    if len(data_x3)>=3:
        col_name = data_x3.columns.tolist()
                           
        for i in range(len(col_name)):
            data_col = data_x3[col_name[i]]
            norm_test = shapiro(data_col)[1]
            
            if norm_test < (1-ci):
                res = est_param(data_col.describe()['mean'], data_col.describe()['std'], len(data_col), ci)
            else:
                res = bootstrap(len(data_col), 1000, ci, data_col)
            
            recommended_low.append(res[0])
            recommended_up.append(res[1])
        
        recommended_dict['column'] = col_name
        recommended_dict['for Low'] = recommended_low
        recommended_dict['for Up'] = recommended_up
        return pd.DataFrame.from_dict(recommended_dict)
```
