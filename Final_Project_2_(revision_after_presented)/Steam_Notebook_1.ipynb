```python
def create_season(cols):
    if cols.month in [3,4,5]:
        return 'Spring'
    elif cols.month in [6,7,8]:
        return 'Summer'
    elif cols.month in [9,10,11]:
        return 'Fall'
    else:
        return 'Winter'
```

```python
def merge_list(col_name):
    # using list comprehension to concat all of categories value into one list
    res = [z for y in [x for x in steam2[col_name]] for z in y] 
    return res
```

```python
platforms_list = merge_list('platforms')
platforms_list
```

```python
def pct_object(object_list):
    unique_obj = pd.Series(object_list).value_counts()
    pct_obj = {}
    
    for i in unique_obj.index:
        pct_obj[i] = unique_obj.loc[i]/len(object_list)

    return pct_obj
```

```python
pct_platforms = pct_object(platforms_list)
pct_platforms
```

```python
def sum_pct(cell_object, pct_object):
    res_sum = 0.0
    for i in cell_object:
        res_sum += pct_object[i]
    
    return res_sum
```

```python
steam2['pct_platforms'] = steam2['platforms'].apply(lambda x: sum_pct(x, pct_platforms))
steam2[['platforms', 'pct_platforms']]
```

```python
categories_list = merge_list('categories')
genres_list = merge_list('genres')
steamspy_tags_list = merge_list('steamspy_tags')
```

```python
pct_categories = pct_object(categories_list)
pct_genres = pct_object(genres_list)
pct_steamspy_tags = pct_object(steamspy_tags_list)
```

```python
steam2['pct_categories'] = steam2['categories'].apply(lambda x: sum_pct(x, pct_categories))
steam2['pct_genres'] = steam2['genres'].apply(lambda x: sum_pct(x, pct_genres))
steam2['pct_steamspy_tags'] = steam2['steamspy_tags'].apply(lambda x: sum_pct(x, pct_steamspy_tags))
```

```python
def desc_filter(x):
    re1 = re.sub('\s+', ' ', x)
    p = re.compile(r'<.*?>')
    
    return p.sub('', re1)
```
