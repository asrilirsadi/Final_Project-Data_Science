```python
def mid_range(x):
    split_x = x.split('-')
    
    diff = int(split_x[1]) - int(split_x[0])
    midrange = int(split_x[0]) + diff/2
    
    return int(midrange)
```

```python
steam3['low_lim_owners'] = steam3['owners'].apply(lambda x: x.split('-')[0])
steam3['up_lim_owners'] = steam3['owners'].apply(lambda x: x.split('-')[1])
steam3['mid_range_owners'] = steam3['owners'].apply(lambda x: mid_range(x))
```
