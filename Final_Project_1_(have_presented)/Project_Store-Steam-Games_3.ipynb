### Inspecting if our residual following the normal distribution or not using Shapiro test

```python
from scipy.stats import shapiro, skew, kurtosis
```

```python
def try_normal(data, residual):
    result = {'Shapiro' : shapiro(data)[1],
              'Skewness': skew(residual),
              'Kurtosis': kurtosis(residual)
    }
#     result = {'Anderson' : {i:j for i,j in zip(anderson(data)[2], anderson(data)[1])},
#              'Shapiro': shapiro(data)[1],
#              'K^2': normaltest(data)[1]}
#     result['Anderson']['stat'] = anderson(data)[0]
    return result
```

```python
try_normal(data, result_low.resid)
```

```python
try_normal(data, result_up.resid)
```

Because the shapiro result test from result1.resid and result2.resid is 0 and it is below than alpha = 0.05, so we can conclude that the residual from result1 and result2 is not Normal Distribution

### Checking for 'Autocorrelation' using Ljungbox test

- H0: Autocorrelation is absent
- H1: Autocorrelation is present

HO is accepted if alpha < p-value

```python
from statsmodels.stats import diagnostic
```

```python
# Autocorrelation checking for result_low.resid

diagnostic.acorr_ljungbox(result_low.resid, lags = 1)
#output (stat, p-value)
```

```python
# Autocorrelation checking for result_up.resid

diagnostic.acorr_ljungbox(result_up.resid, lags = 1)
#output (stat, p-value)
```

Because p-value for result_low.resid and result_up.resid is greater than alpha = 0.05, so for those target we can conclude that each residual is autocorrelation.

### Checking for Heteroscedascity using Goldfeld Quandt test
- H0: Error terms are homoscedastic
- H1: Erorr terms are Heteroscedascity

HO is accepted if alpha < p-value

```python
import statsmodels.stats.api as sms
from statsmodels.compat import lzip
```

```python
# Heteroscedascity checking for result_low.resid

name = ['F statistic', 'p-value']
test = sms.het_goldfeldquandt(result_low.resid, result_low.model.exog)
lzip(name, test)
```

```python
# Heteroscedascity checking for result_up2.resid

name = ['F statistic', 'p-value']
test = sms.het_goldfeldquandt(result_up.resid, result_up.model.exog)
lzip(name, test)
```

Because p-value for result_low.resid and result_up.resid is below than alpha = 0.05, so for those target we can conclude that each residual is Homoscedastic.

### Inspecting the VIF (Variance Inflation Factor)

A variance inflation factor(VIF) detects multicollinearity in regression analysis. Multicollinearity is when there’s correlation between predictors (i.e. independent variables) in a model.

- 1 = not correlated.
- Between 1 and 5 = moderately correlated.
- Greater than 5 = highly correlated.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = [variance_inflation_factor(x_train_low.values, j) for j in range(x_train_low.shape[1])]
pd.Series(vif, index = x_train_low.columns, name = 'VIF')
```

```python
vif = [variance_inflation_factor(x_train_up.values, j) for j in range(x_train_up.shape[1])]
pd.Series(vif, index = x_train_up.columns, name = 'VIF')
```

```python
filter_publisher = []
r2_low_list = []
r2_up_list = []
result_dict = {}

for i in top20_publisher:
    data = new_steam[new_steam['publisher'] == i]
    target_low = data['low_lim_owners']
    target_up = data['up_lim_owners']
    
    x_train_low5, x_test_low5, y_train_low5, y_test_low5 = train_test_split(data.drop(columns=['publisher', 'low_lim_owners', 'up_lim_owners']), target_low, test_size = 0.3, random_state = 101)
    x_train_up5, x_test_up5, y_train_up5, y_test_up5 = train_test_split(data.drop(columns=['publisher', 'low_lim_owners', 'up_lim_owners']), target_up, test_size = 0.3, random_state = 101)
    
    lm_steam_low5 = LinearRegression()
    lm_steam_up5 = LinearRegression()
    lm_steam_low5.fit(x_train_low5, y_train_low5)
    lm_steam_up5.fit(x_train_up5, y_train_up5)
    
    predictions_low5 = lm_steam_low5.predict(x_train_low5)
    predictions_up5 = lm_steam_up5.predict(x_train_up5)
    
    r2_low5 = metrics.r2_score(y_train_low5, predictions_low5)
    r2_up5 = metrics.r2_score(y_train_up5, predictions_up5)
    
    if (r2_low5 > 0.7) and (r2_low5 < 1) and (r2_up5 > 0.7) and (r2_up5 < 1):
        filter_publisher.append(i)
        r2_low_list.append(r2_low5)
        r2_up_list.append(r2_up5)

result_dict['publisher'] = filter_publisher
result_dict['r2 for low_lim_owners'] = r2_low_list
result_dict['r2 for up_lim_owners'] = r2_up_list
```

```python
def calculate_model(publisher):
    coef_low = {}
    coef_up = {}
    
    data_x2 = new_steam[new_steam['publisher'] == publisher]
    target_low = data_x2['low_lim_owners']
    target_up = data_x2['up_lim_owners']
    
    x_train_low6, x_test_low6, y_train_low6, y_test_low6 = train_test_split(data_x2.drop(columns=['publisher', 'low_lim_owners', 'up_lim_owners']), target_low, test_size = 0.3, random_state = 101)
    x_train_up6, x_test_up6, y_train_up6, y_test_up6 = train_test_split(data_x2.drop(columns=['publisher', 'low_lim_owners', 'up_lim_owners']), target_up, test_size = 0.3, random_state = 101)
    
    lm_steam_low6 = LinearRegression()
    lm_steam_up6 = LinearRegression()
    lm_steam_low6.fit(x_train_low6, y_train_low6)
    lm_steam_up6.fit(x_train_up6, y_train_up6)
    
    coef_low['intercept1'] = lm_steam_low6.intercept_
    coef_up['intercept2'] = lm_steam_up6.intercept_
    
    for i in range(len(x_train_low6.columns)):
        coef_low[x_train_low6.columns[i]] = lm_steam_low6.coef_[i]
    for i in range(len(x_train_up6.columns)):
         coef_up[x_train_up6.columns[i]] = lm_steam_up6.coef_[i]
            
    return [coef_low, coef_up]
```

```python
def get_model(publisher):
    coef_low = calculate_model(publisher)[0]
    coef_up = calculate_model(publisher)[1]
    y_formula_low = ''
    y_formula_up = ''
    
    for i,j in zip(coef_low.keys(), coef_low.values()):
        if (i == 'intercept1'):
            y_formula_low = 'Y_low = {:.2f}'.format(j)
        else:
            if str(j)[0] == '-':
                y_formula_low += ' - {:.2f}*X_{}'.format(float(str(j)[1:]), i)
            else:
                y_formula_low += ' + {:.2f}*X_{}'.format(j,i)
    
    for i,j in zip(coef_up.keys(), coef_up.values()):
        if (i == 'intercept2'):
            y_formula_up = 'Y_up = {:.2f}'.format(j)
        else:
            if str(j)[0] == '-':
                y_formula_up += ' - {:.2f}*X_{}'.format(float(str(j)[1:]), i)
            else:
                y_formula_up += ' + {:.2f}*X_{}'.format(j,i)
    
    print(y_formula_low)
    print('\n')
    print(y_formula_up)
```

```python
from scipy import stats
import math
```

```python
def est_param(mean_x, std_x, len_x, ci):
    if len_x < 30:
        critical_value = stats.t.ppf(ci, len_x-1)
    else:
        critical_value = stats.norm.ppf(ci)

    margin_of_error = critical_value*(std_x/math.sqrt(len_x))
    lower = mean_x-margin_of_error
    upper = mean_x+margin_of_error
    
    return [int(round(lower)), int(round(upper))]
```

```python
def bootstrap(n_size, n_boot, ci, data):
    mean_list = list()
    
    for i in range(n_boot):
        resamp_data = data.sample(n = n_size, replace=True)
        mean_resamp = resamp_data.mean()
        mean_list.append(mean_resamp)
    
    a = (1.0-ci)/2.0
    
    #confidence interval
    p1 = (a)*100
    lower = np.percentile(mean_list,p1)
    p2 = (ci+(a))*100
    upper = np.percentile(mean_list,p2)
    
    return [int(round(lower)), int(round(upper))]
```

```python
def recommended_val(publisher, ci):
    data_x2 = new_steam[new_steam['publisher'] == publisher]
    data_x2 = data_x2.drop(columns=['publisher', 'low_lim_owners', 'up_lim_owners'])
    recommended_low = []
    recommended_up = []
    recommended_dict = {}
    
    if len(data_x2)>=3:
        col_name = data_x2.columns.tolist()
                           
        for i in range(len(col_name)):
            data_col = data_x2[col_name[i]]
            norm_test = shapiro(data_col)[1]
            
            if norm_test < (1-ci):
                res = est_param(data_col.describe()['mean'], data_col.describe()['std'], len(data_col), ci)
            else:
                res = bootstrap(len(data_col), 1000, ci, data_col)
            
            recommended_low.append(res[0])
            recommended_up.append(res[1])
        
        recommended_dict['column'] = col_name
        recommended_dict['for Low'] = recommended_low
        recommended_dict['for Up'] = recommended_up
        return pd.DataFrame.from_dict(recommended_dict)
#         print(res_dict)
```
