```python
def correlation_ratio(categories, measurements):
    fcat, _ = pd.factorize(categories)
    cat_num = np.max(fcat)+1
    y_avg_array = np.zeros(cat_num)
    n_array = np.zeros(cat_num)
    for i in range(0,cat_num):
        cat_measures = measurements[np.argwhere(fcat == i).flatten()]
        n_array[i] = len(cat_measures)
        y_avg_array[i] = np.average(cat_measures)
    y_total_avg = np.sum(np.multiply(y_avg_array,n_array))/np.sum(n_array)
    numerator = np.sum(np.multiply(n_array,np.power(np.subtract(y_avg_array,y_total_avg),2)))
    denominator = np.sum(np.power(np.subtract(measurements,y_total_avg),2))
    if numerator == 0:
        eta = 0.0
    else:
        eta = np.sqrt(numerator/denominator)
    return eta
```

```python
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x,y)
    chi2 = ss.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2/n
    r,k = confusion_matrix.shape
    phi2corr = max(0, phi2-((k-1)*(r-1))/(n-1))
    rcorr = r-((r-1)**2)/(n-1)
    kcorr = k-((k-1)**2)/(n-1)
    return np.sqrt(phi2corr/min((kcorr-1),(rcorr-1)))
```

```python
num_columns = []

def corr_ratio_test(choosen_col):
    corr_ratio = []
    
    for i in steam.select_dtypes(exclude = 'object').columns:
        if i != 'appid':
            num_columns.append(i)
            
            corr_ratio.append(correlation_ratio(steam[choosen_col], steam[i])) 

    corr_ratio_pd = pd.DataFrame(data = [num_columns, corr_ratio]).T 
    corr_ratio_pd.columns = ['Column', 'Corr. Ratio Result']
    
    return corr_ratio_pd.sort_values(by = 'Corr. Ratio Result')
```

```python
from scipy.stats import shapiro, anderson, skew, kurtosis
```

```python
alpha = 0.05

normal_test = []
skewness_list = []
kurtosis_list = []

for i in num_columns:       
    if shapiro(steam[i])[1] < alpha:
        normal_test.append('Not-Normal Distribution')
    else:
        normal_test.append('Normal Distribution')

    skewness_list.append(skew(steam[i]))
    kurtosis_list.append(kurtosis(steam[i]))
    
norm_pd = pd.DataFrame(data = [num_columns, normal_test, skewness_list, kurtosis_list]).T 
norm_pd.columns = ['Numerical Column', 'Normality Test', 'Skewness', 'Kurtosis']
norm_pd
```
